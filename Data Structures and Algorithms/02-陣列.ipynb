# **一維陣列**

## **宣告一維陣列**
1. 在電腦程式中使用陣列，都必須需先宣告陣列變數，指定這個陣列的變數名稱
2. 指定陣列的元素個數
3. 陣列初始化：設定陣列每個元素的值，在程式中指定陣列元素的值

以下是設定一個元素個數為5，分別是abcde五個字母的陣列，在Python中要如何撰寫？

圖片1.png

````python
# 宣告一個陣列並指定陣列名稱，這裡我們用Python的List來做範例
array_list = []

# 指定陣列的元素個數
size = 5

# 陣列初始化，初始值都為None，陣列就像是：['None', 'None', 'None', 'None', 'None']
array_list = [None] * size

# 逐一賦值
array_list[0] = 'a'
array_list[1] = 'b'
array_list[2] = 'c'
array_list[3] = 'd'
array_list[4] = 'e'

print(array_list)  # 輸出: ['a', 'b', 'c', 'd', 'e']
````

當然在Python中不需要這麼麻煩，你可以很簡潔設定一個陣列：

````python
# 創建一個一維列表
one_dimensional_list = [1, 2, 3, 4, 5]
````

## **讀取一維陣列**

* 讀取單一資料：可以指定陣列索引值讀取元素
* 一次輸出所有資料：在Python中可以直接列印
* 逐一讀取所有資料：利用迴圈可以逐一存取陣列中所有元素

````python
# 宣告一個一維列陣列
one_dimensional_list = [1, 2, 3, 4, 5]

# 讀取單一資料
print("讀取單一資料：")
print(one_dimensional_list[0])  # 輸出: 1

# 一次輸出所有資料
print("一次輸出所有資料")
print(one_dimensional_list)  # 輸出: [1, 10, 3, 4, 5]

# 逐一讀取所有資料
print("逐一讀取所有資料")
for item in one_dimensional_list:
    print(str(item))

print("也可以這樣寫")
for i in range(5):
    print(str(one_dimensional_list[i]))
````

## **將一個新值插入到一維陣列**

1. 在陣列最後面增加一個空位
2. 從要插入的索引值開始，將所有的資料全部往右移
3. 以插入索引值位置的原有值，更新為新值

````python
# 宣告一個一維列陣列
one_dimensional_list = [1, 2, 3, 4, 5]

# 在陣列最後面增加一個新的數值「0」，
one_dimensional_list.append(0)
print(one_dimensional_list) # 輸出: [1, 2, 3, 4, 5, 0]

# 將6插入到2與3之間，也就是插入索引值為2，要插入一個數值6
# 首先每個元素向右移一格，這裡的range(5, 2, -1)是指，從索引值5迭代-1，直到2為止
for i in range(len(one_dimensional_list)-1, 2, -1):
    one_dimensional_list[i] = one_dimensional_list[i-1]

# 將插入索引值位置2的原有值，更新為新值6
one_dimensional_list[2] = 6

print(one_dimensional_list)
````

以上只是要跟同學說明陣列插入新值的原理是甚麼？

但是在Python做相同的事，其實不用這麼麻煩，可以使用insert方法，將要插入的新值直接指定索引值。

````python
one_dimensional_list = [1, 2, 3, 4, 5]

one_dimensional_list.insert(2, 6)

print(one_dimensional_list)
````

## **刪除一維陣列的特定數值**

1. 從要刪除資料的索引值開始，將所有的資料全部往左移
2. 將陣列的最後一個值刪除

````python
# 宣告一個一維列陣列
one_dimensional_list = [1, 2, 3, 4, 5]

# 如果要刪除2這個數值，從數值2的索引值1開始，將所有的資料全部往左移
# 這裡的range(1, 4)是指，從索引值1開始到索引值4，一一將數值更新為下一個索引值的數值
for i in range(1, len(one_dimensional_list)-1):
    one_dimensional_list[i] = one_dimensional_list[i+1]

# 將插入索引值位置2的原有值，更新為新值6
one_dimensional_list.pop(-1)

print(one_dimensional_list)
````

同樣的，以上只是要跟同學說明刪除一維陣列的特定數值的原理是甚麼？

在Python做相同的事，也不用這麼麻煩，也就是用pop這個方法，將指定索引值的數值刪除。

````python
one_dimensional_list = [1, 2, 3, 4, 5]

one_dimensional_list.pop(1)

print(one_dimensional_list)
````

## **複製一維陣列**

1. 先宣告陣列變數，指定要複製的新陣列的變數名稱
2. 指定新陣列的元素個數
3. 陣列初始化：設定陣列每個元素的值，在程式中指定陣列元素的值
4. 用迴圈讀取舊陣列，將值一一讀出，並且存入新陣列

````python
# 宣告一個陣列並指定陣列名稱，這裡我們用Python的List來做範例
old_one_dimensional_list = [1, 2, 3, 4, 5]
new_one_dimensional_list = []

# 指定陣列的元素個數
size = 5

# 陣列初始化，初始值都為None，陣列就像是：['None', 'None', 'None', 'None', 'None']
new_one_dimensional_list = [None] * size

# 用迴圈讀取舊陣列，將值一一讀出，並且存入新陣列
for i in range(5):
    new_one_dimensional_list[i] = old_one_dimensional_list[i]

print(old_one_dimensional_list)
print(new_one_dimensional_list)
````

同樣的，以上只是要跟同學說明複製一維陣列的原理是甚麼？

在Python做相同的事，也不用這麼麻煩，直接宣告變數指定特定陣列變數就好了。

````python
old_one_dimensional_list = [1, 2, 3, 4, 5]
new_one_dimensional_list = old_one_dimensional_list

print(old_one_dimensional_list)
print(new_one_dimensional_list)
````

# **程式設計為什麼要用陣列？**

假如有一個同學A的期末成績，用Python來表示，你需要一一建立每一科的變數，並且存入資料。

````python
# 同學A的期末成績
math = 90
english = 85
chinese = 70
chemical = 90
physics = 85
````

上述的程式碼的確是可行的，如果要計算所有的期末成績平均，可以用以下的程式碼來計算。

````python
# 計算平均數
average = (math + english + chinese + chemical + physics) / 5
print("同學A的期末成績平均數：" + str(average))
````

不過假如我們要增加音樂的成績，程式就要修改成這樣：

````python
# 同學A的期末成績
math = 90
english = 85
chinese = 70
chemical = 90
physics = 85
music = 60

# 計算平均數
average = (math + english + chinese + chemical + physics + music) / 6

print("同學A的期末成績平均數：" + str(average))
````

因此這樣的程式就會很不好寫，要修改的話也要增加新的變數，假如要計算是全班的成績，光是新增變數就要花很多時間寫程式。這時我們可以把同學A的成績存在陣列裡面，這邊我們使用Python的清單(List)資料結構。

````python
# 同學A的期末成績，以陣列儲存
scores = [90, 85, 70, 90, 85]

# 計算平均數
total_scores = 0
for score in scores:
    total_scores += score
average = total_scores / len(scores)

print("同學A的期末成績平均數：" + str(average))
````

如果要新增一個科目的成績，只要在陣列裡面，直接新增數值就好。

````python
# 同學A的期末成績，以陣列儲存
scores = [90, 85, 70, 90, 85, 60]

# 計算平均數
total_scores = 0
for score in scores:
    total_scores += score
average = total_scores / len(scores)

print("同學A的期末成績平均數：" + str(average))
````

假如要計算全班成績總平均，我們可以使用二維陣列，你只要這樣修改程式。

不管之後需要新增多少同學多少科的成績，你只需要維護那個二維陣列變數的內容即可，不需要修改程式內容。

````python
# 全班成績，每一行代表一位同學的成績，每一列則對應一門科目的成績
students = [[44, 47, 64, 67, 67,  9],
            [83, 21, 36, 87, 70, 88],
            [88, 12, 58, 65, 39, 87],
            [46, 88, 81, 37, 25, 77]]

# 計算成績的函式
def calculate_average(students):
    numbers = 0
    total_scores = 0
    for student_scores in students:
        for score in student_scores:
            total_scores += score
            numbers += 1
    average = total_scores / numbers

    return average

print("全班成績總平均數：" + str(calculate_average(students)))
````

以下是可能的程式碼，你可以想想看，有沒有更好的寫法？

````python
# 全班成績的資料，每一行代表一位同學的成績，每一列則對應一門科目的成績
students_names = ["蛋堡", "陳奕迅", "韋禮安", "周興哲"]
students_scores = [[44, 47, 64, 67, 67,  9],
                   [83, 21, 36, 87, 70, 88],
                   [88, 12, 58, 65, 39, 87],
                   [46, 88, 81, 37, 25, 77]]
students_average_scores = []

# 計算成績的函式
def calculate_individual_average(students_scores):
    average_scores = []

    for scores in students_scores:
        total_temp_score = 0
        for score in scores:
            total_temp_score += score
        average = total_temp_score / len(scores)
        average_scores.append(average)

    return average_scores

# 計算每一個同學的期末成績平均，並且存到陣列裡面
students_average_scores = calculate_individual_average(students_scores)

# 列印所有同學的姓名與期末成績平均
for i in range(len(students_names)):
    print(students_names[i] + "的期末平均分數為：" + str(students_average_scores[i]))
````
