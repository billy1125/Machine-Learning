# **程式設計的本質**

程式設計是資料結構與演算法的結合，以下方程式為範例：

* 資料結構：list列表

```
numbers = [12, 50, 30, 57, 99]
```
* 演算法：則是首先設定 max_value 為列表的第一個元素，然後讀取list列表中的每一個元素。如果遇到一個比 max_value 更大的元素，就更新 max_value。最後列印出 max_value。

資料結構會影響演算法的設計，進而影響程式功能和效率

````python
# 初始化列表
numbers = [12, 50, 30, 57, 99, 1500]

# 假定列表中第一個為最大的數—個元素
max_value = numbers[0]

# 遍歷列表中的每—個元素
for number in numbers:
    # 如果當前元素比知目前最大值大
    if number > max_value:
        # 更新最大值
        max_value = number

# 打印最大值
print("最大值是:", max_value)
````
